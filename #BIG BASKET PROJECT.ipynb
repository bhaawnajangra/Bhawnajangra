```python
MRP= df.groupby('category')['market_price'].sum()
SRP =df.groupby('category')['sale_price'].sum()
(MRP-SRP)/MRP * 100
```

```python
MRP = pd.DataFrame(MRP).reset_index()
SRP = pd.DataFrame(SRP).reset_index()
Dis = pd.DataFrame({'category':MRP.category, 'Original Price':MRP.market_price,
                    'Sale Price':SRP.sale_price, 'Discount Price':MRP.market_price-SRP.sale_price,
                    'Disount Percent':(MRP.market_price-SRP.sale_price)/MRP.market_price*100
})
```

```python
Dis
```

```python
# sactterplot--

plt.figure(figsize=(10,6))
a =sns.scatterplot(x=Dis.category,y=Dis['Disount Percent'],color='#2A11DF',hue=Dis.category,palette='bright')
plt.title('Category wise Discount',fontsize='x-large',fontweight='bold')
plt.xticks(rotation='vertical')
plt.grid(color='blue',linestyle='dashdot')
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix for numeric variables
correlation_matrix = df[numeric_columns].corr()

# Heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap ='PiYG', fmt=".2f")
plt.title('*Correlation Matrix for Numeric Variables*',fontweight='bold')
plt.show()
```
